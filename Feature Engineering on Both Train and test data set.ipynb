# Train Data Missing Value Treatment

```python
# 1 -step make the list of features which has missing values
features_with_na_train=[features for features in train.columns if train[features].isnull().sum()>0]

# 2- step print the feature name and the percentage of missing values
for feature in features_with_na_train:
    print("There is ", np.round(train[feature].isnull().sum()/train.shape[0]*100, 2),'% missing values in',feature)
```

First lets handle Categorical features which has missing values in train data

```python
categorical_features_nan_train=[feature for feature in train.columns if train[feature].isnull().sum()>0 and 
                                train[feature].dtypes=='O']

for feature in categorical_features_nan_train:
    print("There is ", np.round(train[feature].isnull().sum()/train.shape[0]*100, 2),'% missing values in',feature)
```

```python
# Replace missing value with a new label
def replace_cat_feature(train,categorical_features_nan_train):
    data=train.copy()
    data[categorical_features_nan_train]=data[categorical_features_nan_train].fillna('Missing')
    return data

train=replace_cat_feature(train,categorical_features_nan_train)

train[categorical_features_nan_train].isnull().sum()
```

Now lets check for numerical variables that contains missing values in train data

```python
numerical_with_nan_train=[feature for feature in train.columns if train[feature].isnull().sum()>0 and train[feature].dtypes!='O']

# print the numerical nan variables and percentage of missing values
for feature in numerical_with_nan_train:
     print("There is ", np.round(train[feature].isnull().sum()/train.shape[0]*100, 2),'% missing values in',feature)
```

```python
# Replacing the numerical Missing Values

for feature in numerical_with_nan_train:
    # We will replace by using median since there are outliers
    median_value=train[feature].median()
    train[feature].fillna(median_value,inplace=True)
    
train[numerical_with_nan_train].isnull().sum()
```

# Create New Features

Take only number of years from Date Time Variables('YearBuilt', 'YearRemodAdd', 'GarageYrBlt') by subtraction 'yrsold'

```python
# Train data
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    train[feature]=train['YrSold'] - train[feature]

# Test data
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    test[feature]=test['YrSold']-test[feature]
```

Calculating total square feet (area)

```python
# Train data
train['Total_SF'] = train['TotalBsmtSF'] + train['GrLivArea']
train['TotalFloorSF'] = train['1stFlrSF'] + train['2ndFlrSF']
train['TotalPorchSF'] = train['OpenPorchSF'] + train['EnclosedPorch'] + train['3SsnPorch'] + train['ScreenPorch']
train['TotalBathrooms'] = train["FullBath"] + 0.5 * train["HalfBath"] + train["BsmtFullBath"] + 0.5 * train["BsmtHalfBath"]

# Test data
test['Total_SF'] = test['TotalBsmtSF'] + test['GrLivArea']
test['TotalFloorSF'] = test['1stFlrSF'] + test['2ndFlrSF']
test['TotalPorchSF'] = test['OpenPorchSF'] + test['EnclosedPorch'] + test['3SsnPorch'] + test['ScreenPorch']
test['TotalBathrooms'] = test["FullBath"] + 0.5 * test["HalfBath"] + test["BsmtFullBath"] + 0.5 * test["BsmtHalfBath"]
```

Create some boolean features (Yes-No type)

```python
# Train data
train['HasBasement'] = train['TotalBsmtSF'].apply(lambda x: 1 if x>0 else 0)
train['HasGarage'] = train['GarageArea'].apply(lambda x: 1 if x>0 else 0)
train['HasFireplace'] = train['Fireplaces'].apply(lambda x: 1 if x > 0 else 0)
train['HasPorch'] = train['TotalPorchSF'].apply(lambda x: 1 if x>0 else 0)
train['HasPool'] = train['PoolArea'].apply(lambda x: 1 if x>0 else 0)
train['WasRemodeled'] = (train.YearRemodAdd != train.YearBuilt).astype(np.int64)
train['WasCompleted']=train['SaleCondition'].apply(lambda x: 1 if x=='Partial' else 0)
train['HasPavedDrive'] = train['PavedDrive'].apply(lambda x: 0 if x=='N' else 1)

# Test data
test['HasBasement'] = test['TotalBsmtSF'].apply(lambda x: 1 if x>0 else 0)
test['HasGarage'] = test['GarageArea'].apply(lambda x: 1 if x>0 else 0)
test['HasFireplace'] = test['Fireplaces'].apply(lambda x: 1 if x > 0 else 0)
test['HasPorch'] = test['TotalPorchSF'].apply(lambda x: 1 if x>0 else 0)
test['HasPool'] = test['PoolArea'].apply(lambda x: 1 if x>0 else 0)
test['WasRemodeled'] = (test.YearRemodAdd != test.YearBuilt).astype(np.int64)
test['WasCompleted']=test['SaleCondition'].apply(lambda x: 1 if x=='Partial' else 0)
test['HasPavedDrive'] = test['PavedDrive'].apply(lambda x: 0 if x=='N' else 1)
```

```python
Ordinal_features=['GarageType','LotConfig','Street','Alley','Fence','PoolQC','PavedDrive','GarageCond','GarageQual',
                  'GarageFinish','FireplaceQu','Functional','KitchenQual','Electrical','CentralAir','HeatingQC','BsmtFinType2',
                  'BsmtFinType1','BsmtExposure','BsmtCond','BsmtQual','ExterCond','ExterQual','LandSlope','Utilities','LotShape']
```

Remaining categorical features which will be encoded 

```python
# Train data
encode_cat_feature_train=[feature for feature in categorical_features_train if feature not in Ordinal_features + 
                          ['OverallQual','OverallCond']]

# Test data
encode_cat_feature_test=[feature for feature in categorical_features_test if feature not in Ordinal_features +
                         ['OverallQual','OverallCond']]
```

Remove categorical variables that are present less than 1% of the observations

```python
# Train data
for feature in encode_cat_feature_train:
    temp=train[feature].value_counts(normalize=True)
    temp_train=temp[temp>0.01].index
    train[feature]=np.where(train[feature].isin(temp_train),train[feature],'Rare_var')

# Test data
for feature in encode_cat_feature_test:
    temp=test[feature].value_counts(normalize=True)
    temp_test=temp[temp>0.01].index
    test[feature]=np.where(test[feature].isin(temp_test),test[feature],'Rare_var')
```

Categorical fetures target encoding

```python
# Target Encoding
encoder = target_encoder.TargetEncoder(cols = encode_cat_feature_train, smoothing = 0)

# Fit and transform train data
after_encoding_train=encoder.fit_transform(train[encode_cat_feature_train], train['SalePrice'])

# Transform test data
after_encoding_test=encoder.transform(test[encode_cat_feature_test])
```
